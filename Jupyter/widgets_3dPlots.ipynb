```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact, interactive, fixed, widgets
plt.rcParams['font.family']='serif'
plt.rcParams['font.weight']='light'
plt.rcParams['font.size']=14
figsize = (12,6)
dataDir = '/Users/sdporzio/Data/'

from sklearn import datasets
iris = datasets.load_iris()
data = iris.data
test = iris.target
color = test
color = np.where(test==0,'red',color)
color = np.where(test==1,'blue',color)
color = np.where(test==2,'green',color)
```

# Widgets
For continuos variables

```python
def look(elevation=10,azimuth=75):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data[:,0],data[:,1],data[:,2],s=20,c=color)
    ax.view_init(elev=elevation, azim=azimuth)
    
w_elev = (0,90,5)
w_azim = (0,90,5)
interact(look, elevation=w_elev, azimuth=w_azim)
```

# Manual update
For slow functions

```python
def look(elevation=10,azimuth=75):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data[:,0],data[:,1],data[:,2],s=50,c=color)
    ax.view_init(elev=elevation, azim=azimuth)
    
w_elev = (0,90,5)
w_azim = (0,90,5)
interact(look, elevation=w_elev, azimuth=w_azim, __manual=True)
```

# Dropdown
For discrete variables in conjuction with dictionaries

```python
def look(elevation=10,azimuth=75):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data[:,0],data[:,1],data[:,2],s=50,c=color)
    ax.view_init(elev=elevation, azim=azimuth)
    
w_elev = {'ground':0,'up':90}
w_azim = {'front':0,'side':90}
interact(look, elevation=w_elev, azimuth=w_azim)
```
