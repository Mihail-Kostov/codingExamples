```python
nBins = 41
cut = 4
xStart = 20
xEnd = 750

attenuation = []
distance = []

for i in range(len(df['Channel'][0])):
    attenuation.append(df['ADC'][0][i]/float(df['EnergyDeposit'][0][i])*1e-6)
    distance.append(df['ReadoutDistance1'][0][i])
attenuation = np.array(attenuation)
distance = np.array(distance)

bins = np.linspace(xStart,xEnd,nBins)
cont = [[] for x in range(nBins)]
hist = np.linspace(xStart,xEnd,nBins)
yerr = np.linspace(xStart,xEnd,nBins)
for i in range(len(bins)):
    for j in range(len(attenuation)):
        if i!=len(bins): rightEdge = bins[-1] + bins[-1] - bins[-2]
        else: rightEdge = bins[i+1]
        if distance[j] > bins[i] and distance[j] < rightEdge:
            if attenuation[j]<cut: cont[i].append(attenuation[j])

for i in range(len(bins)):
    cont[i] = np.array(cont[i])
    if len(cont[i]) > 1:
        hist[i] = np.average(cont[i])
        yerr[i] = np.std(cont[i])
    else:
        hist[i] = 0
        yerr[i] = 0

# print hist[-1]
# hist = hist[:-1]
# bins = bins[:-1]
# yerr = yerr[:-1]
# print hist[-1]  
    
fig = plt.figure(figsize=figsize)
plt.plot(distance,attenuation,'ro')
plt.ylim(0,5)
plt.xlim(xStart,xEnd)
    
    
# print hist
fig = plt.figure(figsize=figsize)
plt.errorbar(bins,hist,yerr=yerr,fmt='.',color='C0',lw=2,capsize=3,elinewidth=1)
plt.xlim(xStart,xEnd)
plt.ylim(0,2)
```

```python
def func(x,a,b,q0):
    result = a/np.power((x-b),2.) * 1./q0
    return result

popt, pcov = curve_fit(func, bins, hist)
xf=np.linspace(0,1000,50)


fig = plt.figure(figsize=figsize)
plt.errorbar(bins,hist,yerr=yerr,fmt='.',color='C0',lw=2,capsize=3,elinewidth=1)
plt.plot(xf, func(xf, *popt), 'r-', label="Fitted Curve")
plt.title('Attenuation')
plt.ylabel('ADC / Energy deposit [eV]')
plt.xlabel('Distance from readout end [cm]')
plt.xlim(0,800)
plt.ylim(0,2)
```
